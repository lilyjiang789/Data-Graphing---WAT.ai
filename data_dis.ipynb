```python
import matplotlib.pyplot as plt
import numpy as np
import glob
import scipy.io as sio
import seaborn as sns
```

```python
tdf_fs = glob.glob(train_dir + "\*")

print(tdf_fs)

fc_1 = dict(sio.loadmat(train_dir + r"\t12.2022.04.28.mat"))
```
